```python
n=str(input("숫자를 입력하세요:"))
if n.isdecimal():
  n=int(n)
  if n < 30:
    for i in range(1,n+1):
      for j in range(1,i+1):
        print(j,end="")
      print()
  else:
    print("양이 너무 많습니다")
else:
  print("정수로 입력하세요.")
```
